```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
g=9.81
```

```python
deg_angle=45
angle=(deg_angle/360)*(2*np.pi)
constv0=10
v0h=np.cos(angle)*constv0
v0y=np.sin(angle)*constv0
v0=np.linspace(10,30,5) 
vv0y=np.sin(angle)*v0
```

```python
tlim=(2*v0y)/g
t=np.linspace(0,tlim,100)
```

```python
distance=v0h*t
h=v0y*t-(0.5*g*(t**2))
```

```python
plt.plot(t,h)
plt.show()
print(f"Angle:{round(deg_angle,2)} degrees")
print(f"Initial Velocity:{constv0} m/s")
```
